### Step 1: Import your libraries
Let's start by importing:
1. pandas as pd

```python
# Step 1: Import your library
import pandas as pd
```

### Step 2: Download and load red wine CSV
You can get the red wine data directly from the UCI Machine Learning Repository using the url, and the read_csv :

http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv

Two ways to do this:
1. Click the link and download the CSV and reading the CSV using pandas
2. Using pandas to read the link directly + exporting the CSV into your current working folder with to_csv method

<font color='red'>Make sure you have a copy of the red wine CSV in this in your folder - we'll be using it in Part II later!</font>

```python
# Step 2: Download and load the red wine CSV
df = pd.read_csv("winequality-red.csv", sep = ';')
df
```

### [Troubleshoot] Fix the CSV read
It can be tricky if you aren't familiar with the dataset, or if you read the CSV straightaway without opening them in a file editor/Excel. 

If you obtained something like this:



Don't panic! It's just that while CSV stands for Comma Separated Values, sometimes the values in a CSV file are separated by other symbols. In our case, it's a semicolor (";"). 

pandas's read_csv has an addition parameter you can use if your file happens to have a different separator. Check out the <a href = "https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html">documentation</a>! 

<strong>Hint: Google "how to read a file with a semi colon separator in pandas"</strong>

### Step 3: Add isRed column to red wine DataFrame
We will be combining this red wine data with white wine data, so we need to be able to tell which row is which.

Add a column called isRed where the values are all 1, indicating that the rows are indeed red wine.

<strong>Hint: Google "add new column to Pandas dataframe with default value"</strong>

```python
# Step 3: Add isRed column
df['isRed'] = 1
df
```

### Step 4: Download and load white wine CSV
After we obtain the red wine CSV, we download the white wine CSV at:

http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv

Similarly, you can choose to either download it into your folder first before reading the CSV, or reading it directly with the read_csv method. 

<font color = 'red'>Make sure you have a copy of the white wine CSV exported as well!</font>

```python
# Step 4: Download and load the white wine CSV
df2 = pd.read_csv("winequality-white.csv", sep = ';')
df2
```

### Step 5: Add isRed column to white wine DataFrame
Add a isRed column to the white wine DataFrame, except this time round the value is 0, as 0 means 'no'. 

```python
# Step 5: Add isRed column to white wine DataFrame
df2['isRed'] = 0
df2
```

### Step 6: Merge the two DataFrames
Now that we have both DataFrames, it's time to merge them into one. 

Stack the DataFrames on top of one another. 



You are expected to see:
1. 6,497 rows
2. 13 columns

Useful guide here: https://pandas.pydata.org/docs/user_guide/merging.html

```python
# Step 6: Merge the two DataFrames together
df_comb = pd.concat([df,df2], axis = 0)
df_comb
```

### Step 7: Check combined DataFrame for missing values
Before we export this into a CSV, let's see if there is any missing data.

<strong>Hint: "how to count the NaN values in a column in pandas"</strong>

```python
# Step 7: Check for missing data
df_comb.isnull().sum()
```

### Step 8: Export the combined DataFrame as CSV
There should be no missing data, which is great - it's a good habit to check whenever you work with data.

```python
# Step 8: Export the DataFrame as CSV
df_comb.to_csv("df_comb.csv")
```
